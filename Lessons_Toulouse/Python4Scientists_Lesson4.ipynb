3rd dimension are the layers <br>
layers = [R, G, B, alpha (transparency)]

```python
red = img_array[:, :, 0]
red.shape
```

```python
Image.fromarray(red)
```

```python
copernicus = img_array[160:, :470, :]
Image.fromarray(copernicus)
```

```python
alpha = copernicus[:, :, -1]
alpha
```

The higher the alpha value, the more opaque is the image. Let's set alpha to 100.

```python
alpha = 100
alpha
```

```python
alpha = copernicus[:, :, -1]
alpha[:, :] = 100
alpha
```

```python
Image.fromarray(copernicus)
```

**Note**: Note that in the previous example, we changed the variable (array) alpha and the copernicus has changed as well... that happens because when SLICING, it returns a view of the object and not a copy of the object.<br>
In this case, variable alpha is exactly the last layer of copernicus array. Any change made to alpha, will change copernicus as well.

## Splitting and Concatenation

```python
copernicus_red = copernicus[:, :, 0]
print(copernicus_red.shape)
Image.fromarray(copernicus_red)
```

```python
top, bottom = np.split(copernicus_red, [143], axis=0)
Image.fromarray(top)
```

```python
left, right = np.split(copernicus_red, [235], axis=1)
Image.fromarray(left)
```

```python
concat = np.concatenate([right, left], axis=1)
Image.fromarray(concat)
```

## Functions

A function is a block of code which only runs when it is called.

You can pass data, known as parameters, into a function.

A function can return data as a result.

```python
def square(x):
    return x**2

square(11)
```

```python
def power(base, expoent):
    return base**expoent

power(2, 8)
```

```python
# base and expoent are required arguments
# power(2)
```

```python
power(expoent=8, base=2)
```

```python
# b is an optional (default) parameter
# def y = a*x + b
def linear(x, a, b=0):
    y = a * x + b
    return y

linear(2, 3)
```

```python
# nomenclature 2 is a positional argument
# -10 and 3 are keyword arguments
linear(2, b=-10, a=3)
```

```python
# y = a*x^2 + b*x + c
def quadratic(x, a, b=1, c=0):
    y = a * power(x, 2) + b*x + c
    return y
```

```python
quadratic(a=1, x=2)
```

## The value <b>None</b>


The value **None** is used to define a null value. **It is not the same as an empty string, False, or a zero.** It is a data type of the class NoneType object.

```python
None == "", None == False, None == 0
```

```python
# correct way of testing (is -> identity operator)
x = 0
x is None, x is not None
```

```python
bool(None), bool(0), bool(False), bool(123)
```

```python
# None is absence of value, so we cannot implement this...
def check_value(x):
    if x:
        print(f'{x} has a value')
    else:
        print(f'{x} has no value at all')
```

```python
check_value(None), check_value(0), check_value(False)
```

```python
# Why is that... that's because if x: is testing the BOOLEAN result of x.
# To test it correctly we should do:

def check_value(x):
    if  x is None:
        print(f'{x} has no value at all')
    else:
        print(f'{x} has a value')
```

```python
check_value(None), check_value(0), check_value(False)
```
